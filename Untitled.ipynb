```python
# Import required libraries
import numpy as np
import pandas as pd
import hvplot.pandas
import requests

import warnings
warnings.filterwarnings('ignore')

# Libraries to Extract Stock Price Historical Data 
# from Yahoo! Finance using Tickers from Wikipedia
import yfinance as yf
from pandas.tseries.offsets import DateOffset
from datetime import datetime, timedelta
from pandas.tseries.holiday import USFederalHolidayCalendar

today = datetime.today().strftime('%Y-%m-%d')
```

```python
stock_data = pd.read_csv('predicted_price.csv',index_col='Date',parse_dates=True,infer_datetime_format=True)
stock_14days = stock_data.loc['2024-03-28':]
stock_14days
```

```python
stock_pred_error = pd.read_csv('model_eval.csv')
stock_pred_error = stock_pred_error.rename(columns={'Unnamed: 0':'Stock'})

stock_pred_error
```

```python
# Selecting stocks where prediction errors are below threshold
select_stock_list=list(stock_pred_error.loc[(np.where((stock_pred_error['RMSE']<2.50) &
                                             (stock_pred_error['MAPE']<1.50))),'Stock'])

select_stock_list
```

```python
# Strategy: Over next 14 days of predicted price, 
# buy@ lowest predicted price and 
# sell@ highest predicted price, 
# where the percent gain is higher than risk free rate(5.00%)

strategy_stocks = pd.DataFrame(index=select_stock_list,
                               columns=['buy@','sell@','Prob_Gain','Percent_Gain'])

for stock in select_stock_list:
    high_price = pd.DataFrame(stock_14days.loc[:,stock].sort_values(ascending=False)).iloc[0]
    low_price = pd.DataFrame(stock_14days.loc[:,stock].sort_values(ascending=True)).iloc[0]

    stock_high_low = pd.concat([high_price,low_price],axis=1)
    if stock_high_low.columns[0]>stock_high_low.columns[1]:
        stock_high_low['buy@'] = low_price
        stock_high_low['sell@'] = high_price
        stock_high_low['Prob_Gain'] = round((high_price-low_price),3)
        stock_high_low['Percent_Gain'] = round((stock_high_low['Prob_Gain']/low_price)*100,2)
        stock_high_low = stock_high_low[['buy@','sell@','Prob_Gain','Percent_Gain']]
        strategy_stocks = pd.concat([strategy_stocks,stock_high_low],axis=0).dropna()

strategy_stocks
    
```

```python
strategy_stocks = strategy_stocks[strategy_stocks['Percent_Gain']>5.00]
strategy_stocks
```
